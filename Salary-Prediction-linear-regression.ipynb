```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Salary_Data.csv")
df.head()
```

```python
x=df["YearsExperience"]
y=df["Salary"]
plt.scatter(x,y)
plt.title("Years of Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.30, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
x_train2=np.array(x_train).reshape(-1,1)
lr=LinearRegression()
lr.fit(x_train2,y_train)
#lr.fit(x_train, y_train)
```

```python
x_test2=np.array(x_test).reshape(-1,1)
y_predict=lr.predict(x_test2)
```

```python
x_min, x_max=df.YearsExperience.min(),df.YearsExperience.max()
x_new=pd.DataFrame({"YearsExperience" :[x_min, x_max]})
x_predicted=lr.predict(x_new)
print(x_new)
print(x_predicted)
df.plot(kind="scatter", x="YearsExperience", y="Salary")
plt.plot(x_new, x_predicted, c="green", linewidth=2)
```

```python
# Training set
result_train=plt
result_train.scatter(x_train2, y_train, color='blue')
result_train.plot(x_train2, lr.predict(x_train2), color='green')
result_train.title("Salary VS Experience(Training Set)")
result_train.xlabel("Years of Experience")
result_train.ylabel("Salary")
result_train.show()
```

```python
# Testing set
result_test=plt
result_test.scatter(x_test2, y_test, color='magenta')
#result_test.plot(x_test2, lr.predict(x_test2), color='green') same output
result_test.plot(x_train2, lr.predict(x_train2), color='green')
result_test.title("Salary VS Experience(Testing Set)")
result_test.xlabel("Years of Experience")
result_test.ylabel("Salary")
result_test.show()
```
